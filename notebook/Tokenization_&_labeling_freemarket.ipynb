```python

import re
import pandas as pd

class ProductCatalog:
    def __init__(self, products=None):
        self.products = products or["የልጆች መመገቢያ ጡጦ","ዘመናዊ የልብስ ማስቀመጫ ቁምሳጥን","ዋስ ምጣድ","የመፀዳጃ ቤትዕቃዎች  መደርደሪያ","ግላስ የውሀ ማቅረቢያ","ልብስ ማስጫ","የብረት ፔርሙዝ",
                          "ለወይን እና  ለተለያዩ የመጠጥ ማቅረቢያ እና መጠጫ ብርጭቆዎች","ቦማ  ያለዘይት መጥበሻ","የሕፃን ማዘያ","የቤት መወልወያ","አራት በአንድ የማርብል መጥበሻ","ኦርጂናል የጥብስ ማቅረቢያ",
                          "የጫማ ሽፋን","ፕላስተር","ሁለገብ መፍጫ","የፀጉር ፔስትራ","ቦማ  ያለዘይት መጥበሻ","በኤሌትሪክ የሚሰራ ምጣድ","ሲልቨር ክረስት ተች የኤሌክትሪክ ስቶቭ","የፅዳት ቡሩሽ",
                          "አነስተኛ የልብስ እና የጫማ  ማጠቢያ ማሽን","ፊሊፕስ ካውያ","ሱታይ ተች የኤሌክትሪክ ስቶቭ","የእንቁላል መምቻ እና የኬክ ቅርፅ ማውጫ","የቆሸሹ ልብሶች ማስቀመጫ ቅርጫት","የቤት መወልወያ",
                          "የሸቶ መቀነሻ","ማድረቂያ","ከርል መስርያ","ኦቨን ውስጥ የሚገቡ የመስታወት ፓትራዎች","የእንቁላል መምቻ","የብና  መፍጫ","ባለ ክዳን መጥበሻ","ዲጂታል ሚዛን","የፀጉር መቁረጫ","የቤት ውስጥ የፊት ስቲም",
                          "የፀጉር እስቲመር","የልጆች መመገቢያ ጡጦ","የሕፃን ማዘያ","ምቾት ያለው ምንጣፍ","የጡጦ መቀቀያ","የሚተጣጠፍ ማጠቢያ","እናት ምጣድ","የሱሪ ማስቀመጫ","ዘመናዊ የፊት ማፅጃ",
                          "የዉሀ ኮዳ ሴት","የፍሳሽ ማስወገጃ ቱቦዎችን ማፅጂያ ኬሚካል","የፅዳት ቡሩሽ","ሳቺ ባለ ሁለት ጆግ የጁስ መፍጫ ","የኤሌክትሪክ ስቶቭ","የመፀዳጃ ቤትዕቃዎች  መደርደሪያ","ተንቀሳቃሽ ሻወር",
                          "የውሃ ማፍያ","የንግስተ ሳባ ምርቶች","ጀበና","12 ስኒ ከማስቀመጫዎቹ","6ማንኪያ","የወተት ማቅረቢያ","የሰኳር ማቅረቢያ","የገበታ ምንጣፍ","ሳባ ሴራሚክ የወጥ ማቅረቢያ","የንግስተ ሳባ መመገቢያ ሳህን",
                          "የእንጀራ እና ዳቦ ማቅረቢያ","ካሲዮ 5 ሴት የወጥ ቤት እቃዎች","የቁም ስቲም ካውያ","","ፊሊፕስ ካውያ","የሱሪ ማስቀመጫ","የጡት ማለቢያ","አነስተኛ የልብስ እና የጫማ  ማጠቢያ ማሽን",
                          "የእግር ማሳጀር","ለልጆ ጤንነት ወሳኝ የጡጦ መቀቀያ","የእግር ማሳጀር","የቡና እና የቅመማ ቅመም መፍጫ","ሶኒፈር የቡና መቁያ","የአዋቂ ዳይፕር","የሕፃን ማዘያ","የጣሊያን ኦርጅናል ኮንፈርት",
                          "የፅጉር መቁረጫ ማሽን","የሚተጣጠፍ የሚዘረጋ ጆግ","5 አይነት መለዋወጫ ያለው ሁለገብ መፍጫ","የኤሌክትሪክ ጋቻ ","ለረመዳን የቤት  ማድመቂያ  ጌጦች","የጥራጥሬ ማስቀመጫ","ስማርት ተንቀሳቃሽ ሻወር",
                          "የወጥ ቤት እቃዎች ማድረቂያ","ስድስት ስኒ ከነማስቀመጫው","ዲጂታል ሚዛን","ዘመናዊ ገመድ አልባ ጸጉር ከርል መስሪያ","የፍሪጅ ማስቀመጫ እና ማቀሳቀሻ","ሳቺ የሊጥ ማቡኪያ እና መቀላቀያ",
                          "የልብስ ማስጫ","የፅዳት ቡሩሽ"," ገመድ አልባ ጸጉር ከርል መስሪያ","የስፖርት ኮዳ","የመነፅር  ስራ","የሚተጣጠፍ የሚዘረጋ ጆግ","የእግርዎን ልስላሴ ለመጠበቅ"," የለብስ እና የጨማ መደርደሪያ",
                          "ኤሌክትሪክ ግሪል","የሽንኩርት እና የስጋ መፍጫ ማሽን","የቤት ውስጥ የፊት ስቲም","የእግር ማሳጀር","የመፀዳጃ ቤትዕቃዎች  መደርደሪያ","ኤሌክትሪክ ግሪል","የመጠጥ ዉሃ የሚስብ ተንቀሳቃሽ ማሽን",
                          "በራሱ የሚቀላቅል ኩባያ","ሳዮና እንፋሎት ካውያ","የሕጻናት አጫዋች","ሳባ ሴራሚክ የወጥ ማቅረቢያ","የሽንኩርት እና የስጋ መፍጫ ማሽን","የህፃናታ ገላ መታጠቢያ ገንዳ"," የማይዙ ድስቶች",
                          "የድግስ ጣፋጭ መጠጦች ማቅረቢያ","ጂፓስ የስጋ መፍጫ","ምንጣፍ ከወለል ጋር አጣብቆ የሚይዝ","አዲስ ዘመናዊ የፊት ማፅጃ","ተጣጣፊ ፑሽ አፕ መሳሪያ",
                          "ሴት እጅግ በጣም ኳሊቲ እና ዘመናዊ የባኛው ቤት እቃ","የሳሙና ማስቀመጫ","የልጆች ጥርስ መቦረሻ","ልብስ ማስጫ","አመቺና ቀላል ጽዳት ማጠቢያ ጓንት","በቆሻሻ የተዘጉ የፍሳሽ ማስወገጃ ቱቦዎችን ማፅጂያ ኬሚካል","ሳቺ የሊጥ ማቡኪያ እና መቀላቀያ",
                          "ሳቺ የቡና መቁያ","የአምፖል ሴኪውሪቲ ካሜራ","5 አይነት መለዋወጫ ያለው ሁለገብ መፍጫ","የማፍን ኬክ መስሪያ/መጋገሪያ ፓትራ","የሚተጣጠፍ የሚዘረጋ ጆግ","ለሕፃን ወተት ማሞቂያ",
                          "የፀሐይ መነፅር ገመድ አልባ ብሉቱዝ","የኤሌክትሪክ ማሞቂያ ያለው የምሳ እቃ","የመስታዎት ፓትራዎች","ባሎት መጥበሻ በቀላሉ ፓንኬኮችን መስሪያ","የማፍን ኬክ መስሪያ","የፀሐይ መነፅር ገመድ አልባ ብሉቱዝ"
]

        self.b_location = ['ፒያሳ ጣይቱ ሆቴል','ጣይቱ','ሆቴል','መተባበር','መተባበር ሕንፃ','መገናኛ','ፒያሳ','ሳር','ሳር ቤት',
                           'ካናዳ','ኤምባሲ','ጀሞ','ከፍደም','ጀሞ መስታወት ፍብሪካ','ግራውድ','ሸዋ','ሱፐር','ማርኬት']


    def search_product(self, query):
        """
        Search for products that match the query string using a regular expression.
        """
        pattern = re.compile(query)
        result = [product for product in self.b_product if pattern.search(product)]
        return result

    def search_location(self, query):
        """
        Search for locations that match the query string using a regular expression.
        """
        pattern = re.compile(query)
        result = [location for location in self.b_location if pattern.search(location)]
        return result

    def label_products(self):
        labeled_products = []
        for product in self.products:
            tokens = product.split()  # Split the product name into tokens
            if tokens:  # Check if there are tokens
                labeled_product = []
                labeled_product.append((tokens[0], 'B-product'))  # Label the first token as B-product
                for token in tokens[1:]:
                    labeled_product.append((token, 'I-PRODUCT'))  # Label remaining tokens as I-PRODUCT
                labeled_products.append(labeled_product)
        return labeled_products


    def label_message_utf8_with_birr(self, messages):
        all_rows = []  # To hold rows of tokens and labels

        # Go through each message in the dataset
        for message in messages:
            labeled_tokens = []
            found_price_start = False  # Tracks if 'ዋጋ' was found
            found_price_end = False  # Tracks if 'ብር' was found

            # Ensure the message is a string, and skip if it's not valid
            if isinstance(message, float):
                message = str(message)  # Convert float to string if needed

            tokens = message.split()  # Split the message into tokens

            for i, token in enumerate(tokens):
                # Add the new rule for handling tokens like '2000ብር' first
                if re.match(r'^\d+ብር$', token):  # Match numbers attached to 'ብር'
                    labeled_tokens.append((token, 'I-PRICE'))  # Label as I-PRICE for tokens like '2000ብር'
                    continue  # Skip the rest of the checks for this token

                # Check if token is part of the product list
                product_found = False
                for product in self.products:
                    product_tokens = product.split()  # Split product into tokens
                    if len(product_tokens) > 0 and token == product_tokens[0]:  # Check if product token is non-empty
                        labeled_tokens.append((token, 'B-PRODUCT'))  # Label as B-PRODUCT
                        # Label remaining product tokens as I-PRODUCT
                        for j in range(1, len(product_tokens)):
                            if i+j < len(tokens) and tokens[i+j] == product_tokens[j]:
                                labeled_tokens.append((tokens[i+j], 'I-PRODUCT'))
                            else:
                                break
                        product_found = True
                        break  # Exit product loop once match is found

                if not product_found:
                    # Check for price pattern starting with 'ዋጋ' and ending with 'ብር'
                    if token == 'ዋጋ' or token == 'ዋጋ፦':  # Match 'ዋጋ' or 'ዋጋ፦'
                        found_price_start = True
                        labeled_tokens.append((token, 'B-PRICE'))  # Start price sequence with B-PRICE
                    elif found_price_start and re.match(r'^\d+$', token):  # If a number follows 'ዋጋ'/'ዋጋ፦'
                        labeled_tokens.append((token, 'I-PRICE'))  # Label as I-PRICE
                    elif found_price_start and token == 'ብር':  # End of price sequence
                        labeled_tokens.append((token, 'I-PRICE'))
                        found_price_start = False  # Reset for next possible sequence
                    else:
                        # Check if token is a price number not part of a 'ዋጋ'-'ብር' sequence
                        if re.match(r'^\d+(\.\d{1,2})?$', token):
                            labeled_tokens.append((token, 'O'))  # Label as O if it's just a number
                        # Check if token could be a location (e.g., cities or general location names)
                        elif any(loc in token for loc in [
                                'ፒያሳ ጣይቱ ሆቴል', 'ጣይቱ', 'ሆቴል', 'መተባበር',
                                'መተባበር ሕንፃ', 'መገናኛ', 'ፒያሳ', 'ሳር',
                                'ሳር ቤት', 'ካናዳ', 'ኤምባሲ', 'ጀሞ',
                                'ከፍደም', 'ጀሞ መስታወት ፍብሪካ', 'ግራውድ',
                                'ሸዋ', 'ሱፐር', 'ማርኬት']):
                            labeled_tokens.append((token, 'I-LOC'))
                        # Assume other tokens are general text
                        else:
                            labeled_tokens.append((token, 'O'))

            # Store tokens and their labels in a single row
            tokens_row = [token for token, label in labeled_tokens]
            labels_row = [label for token, label in labeled_tokens]
            all_rows.append({'Tokens': tokens_row, 'Labels': labels_row})

        # Create a DataFrame from all rows
        df = pd.DataFrame(all_rows)

        return df
messages = cleaned_amharic_text['Cleaned_Message'].tolist()  # Convert the column to a list

catalog = ProductCatalog()
labeled_df = catalog.label_message_utf8_with_birr(messages)
# Display results in a readable format
for index, row in labeled_df.iterrows():
    print(f"Message {index + 1}:")
    for token, label in zip(row['Tokens'], row['Labels']):
        print(f"  Token: {token}, Label: {label}")
    print()  # Blank line between messages
```

```python
# @title
# Function to save the DataFrame in CoNLL format
def save_to_conll(df, filename):
    with open(filename, 'w', encoding='utf-8') as f:
        for index, row in df.iterrows():
            for token, label in zip(row['Tokens'], row['Labels']):
                f.write(f"{token} {label}\n")
            f.write("\n")  # Blank line to separate sentences/messages

# Example usage with your DataFrame
messages = cleaned_amharic_text['Cleaned_Message'].tolist()  # Convert the column to a list
catalog = ProductCatalog()
labeled_df = catalog.label_message_utf8_with_birr(messages)

# Save the labeled DataFrame to a CoNLL formatted file
conll_filename = 'labeled_data_freemarket_final.conll'
save_to_conll(labeled_df, conll_filename)

print(f"Data saved to {conll_filename} in CoNLL format.")
```
